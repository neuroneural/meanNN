```python
def entropyLoss(X):
    d = X.shape[0]
    n = X.shape[1]

    eigen = torch.eye(n, n)
    squared_norms = (X**2).sum(0).repeat(n, 1)
    squared_norms_T = squared_norms.T
    X_T = X.T
    arg = squared_norms + squared_norms_T - 2 * torch.mm(X_T, X)
    expression = torch.sum(torch.log(torch.abs(arg)+ eigen + 1e-12))/2
    entropy = -d/(n*(n-1)) * expression
    return entropy
```
