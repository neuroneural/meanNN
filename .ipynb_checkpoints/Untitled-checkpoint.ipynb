```python
def entropyLoss(X, use_gpu=True):
    if use_gpu:
        entropy = torch.tensor([0]).type(torch.float32).cuda()
    else:
        entropy = torch.tensor([0]).type(torch.float32)
        
    for batch in X:
        d = batch.shape[0]
        n = batch.shape[1]
        
        if use_gpu:
            eigen = torch.eye(n, n).cuda()
        else:
            eigen = torch.eye(n, n)
        squared_norms = (batch**2).sum(0).repeat(n, 1)
        squared_norms_T = squared_norms.T
        batch_T = batch.T
        arg = squared_norms + squared_norms_T - 2 * torch.mm(batch_T, batch)
        expression = torch.sum(torch.log(torch.abs(arg)+ eigen + 1e-12))/2
        entropy += -d/(n*(n-1)) * expression
    return entropy/X.size()[0]
```
